```python
import pandas as pd 
meteo_data = pd.read_csv("/kaggle/input/data-set/Q_01_previous-1950-2022_RR-T-Vent.csv/Q_01_previous-1950-2022_RR-T-Vent.csv",sep=';')
meteo_data.tail(1000)
```

```python
years_to_keep = [str(year) for year in range(2000, 2023)]
filtered_meteo_data = meteo_data[meteo_data['AAAAMMJJ'].astype(str).str[:4].isin(years_to_keep)]

filtered_meteo_data
```

```python
import pandas as pd 
meteo_data_dep02 = pd.read_csv("/kaggle/input/data-set/Q_02_previous-1950-2022_RR-T-Vent.csv/Q_02_previous-1950-2022_RR-T-Vent.csv",sep=';')
meteo_data_dep02.tail(1000)
```

```python
years_to_keep = [str(year) for year in range(2000, 2023)]
filtered_meteo_data_dep02 = meteo_data_dep02[meteo_data_dep02['AAAAMMJJ'].astype(str).str[:4].isin(years_to_keep)]

filtered_meteo_data_dep02
```

```python
meteo_data02 = filtered_meteo_data_dep02.dropna(axis=1, how='all')
meteo_data02 = meteo_data02.dropna(axis=0, how='all')
nan_percentage = meteo_data02.isna().mean() * 100

# supprimer les colonnes avec plus de 50% de NaN
columns_to_drop = nan_percentage[nan_percentage > 60].index
meteo_data02 = meteo_data02.drop(columns=columns_to_drop)
meteo_data02
```

```python
cols_to_drop = ['QRR', 'TN','QTN','TX','QTX','QTNTXM','QTAMPLI']
meteo_data021 =meteo_data02.drop(cols_to_drop, axis=1) 
meteo_data021 =meteo_data021.dropna(axis=0) 
meteo_data021
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#LinearRegression avec temperature 

features = meteo_data021[['NUM_POSTE','LAT', 'LON', 'ALTI', 'AAAAMMJJ', 'TNTXM', 'TAMPLI']]
target = meteo_data021['RR']

# Division des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Création du modèle de régression (par exemple, régression linéaire)
model = LinearRegression()

# Entraînement du modèle
model.fit(X_train, y_train)

# Prédiction sur les données de test
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
print(f"MSE (Mean Squared Error) : {mse}")
```

```python

# Créer un DataFrame de démonstration avec de nouvelles données pour la prédiction
new_data = {'NUM_POSTE':[2031001,2031001],
            'LAT': [49.834000, 49.5],
            'LON': [4.200500, 4.2],
            'ALTI': [177, 170],
            'AAAAMMJJ': [20000101, 20220402],
            'TNTXM': [4.0, 5.0],
            'TAMPLI': [4.0, 4.0]}
new_df = pd.DataFrame(new_data)
# Faire des prédictions sur les nouvelles données
predictions = model.predict(new_df)

# Afficher les prédictions
print("Prédictions de pluie (RR) :")
for i, pred in enumerate(predictions):
    print(f"Date AAAAMMJJ: {new_data['AAAAMMJJ'][i]}, Prédiction RR: {pred}")
```
